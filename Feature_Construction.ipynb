```python
import pandas as pd
import SimpleCV as cv
```

```python
disp = cv.Display(displaytype='notebook')
```

```python
solutions = pd.read_csv("./input_data/training_solutions_rev1.csv", index_col="GalaxyID")
solutions = solutions[["Class1.1", "Class1.2", "Class1.3"]]
```

```python
threshold = 0.6
```

```python
class1_1 = solutions['Class1.1']
strongly_class1_1 = class1_1[class1_1 >= threshold]

class1_2 = solutions['Class1.2']
strongly_class1_2 = class1_2[class1_2 >= threshold]

class1_3 = solutions['Class1.3']
strongly_class1_3 = class1_3[class1_3 >= threshold]

galaxy_id = strongly_class1_2.index[3]
img_path = "./input_data/images_training_rev1/" + str(galaxy_id) + ".jpg"
img = cv.Image(img_path)
img.save(disp)
```

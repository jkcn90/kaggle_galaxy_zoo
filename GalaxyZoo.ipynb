#### Sample Features (Raw)

```python
img_raw = cv.Image("./input_data/images_training_rev1/170018.jpg")
blobs = img_raw.findBlobs()
blobs = blobs.filter(blobs.area() == max(blobs.area()))
largest_blob = blobs[0]

# Rotate
angle = largest_blob.angle()
w = largest_blob.minRectWidth()
h = largest_blob.minRectHeight()

if w < h:
    angle -= 90

img_raw = img_raw.rotate(angle)

# Find Dimensions to Crop
bounding_box_xywh = largest_blob.boundingBox()

center = largest_blob.centroid()
max_dim = max(bounding_box_xywh[2], bounding_box_xywh[3]) * 1.5
xywh = center+(max_dim, max_dim)

# Display Cropped Image
cropped_image = img_raw.crop(xywh, centered=True)
cropped_image.save(disp)
```

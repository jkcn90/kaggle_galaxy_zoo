## Part 0: Explore Data

```python
import os
import pandas as pd
import SimpleCV as cv
from sklearn import cross_validation

from galaxy_data import GalaxyData
import run
```

```python
disp = cv.Display(displaytype='notebook')
```

```python
img = cv.Image("./input_data/images_training_rev1/100008.jpg")
```

```python
img_temp = img.copy()
blobs = img_temp.findBlobs()
blobs.draw()
img_temp.save(disp)
```

```python
img_temp = img.copy()
blobs = img_temp.findBlobs()
blobs = blobs.filter(blobs.area() == max(blobs.area()))
cropped_image = blobs.crop()[0]
cropped_image = cropped_image.toHLS()
cropped_image.save(disp)
```

### Sample Features

```python
feature_area = cropped_image.area()
(feature_hue, feature_lightness, feature_saturation) = cropped_image.meanColor()
feature_array = [feature_area, feature_hue, feature_lightness, feature_saturation]
print(feature_array)
```

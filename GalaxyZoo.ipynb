## Part 0: Explore Data

### Data Statistics

```python
solutions = pd.read_csv("./input_data/training_solutions_rev1.csv", index_col="GalaxyID")
solutions = solutions.ix[:,"Class1.1"]

class0_bound = 0
class1_bound = 1

solutions = solutions[(solutions >= class1_bound) | (solutions <= class0_bound)]
index0 = random.sample(solutions[solutions<=class0_bound].index, solutions[solutions>=class1_bound].size)
index1 = solutions[solutions>=class1_bound].index

index = list(index1)+index0
solutions = solutions[index]
solutions.hist()
```

```python
feature_vector_list = []
for (num, galaxy_id) in enumerate(solutions.index):
    print(num/float(len(solutions)))
    img_path = os.path.join("./input_data/images_training_rev1/", str(galaxy_id) + ".jpg")
    img = cv.Image(img_path)
    
    blobs = img.findBlobs()
    largest_blob = blobs.filter(blobs.area() == max(blobs.area()))
    largest_blob = largest_blob[0]

    cropped_image = blobs.crop()[0]
    cropped_image = cropped_image.toHLS()
    
    feature_aspect_ratio = largest_blob.aspectRatio()
    
    (feature_hue, feature_lightness, feature_saturation) = cropped_image.meanColor()
    feature_vector = [feature_aspect_ratio, feature_hue, feature_saturation]
    feature_vector_list.append(feature_vector)
```

```python
feature_vector_list = pd.DataFrame(feature_vector_list, index=solutions.index)

binary_solutions = solutions.copy()
binary_solutions[solutions>=class1_bound] = 1
binary_solutions[solutions<=class0_bound] = 0

features_and_solutions = feature_vector_list.join(binary_solutions)
```

```python
groups = features_and_solutions.groupby('Class1.1')

(fig, ax) = plt.subplots()
ax.margins(0.05)
for name, group in groups:
    ax.plot(group.icol(0), group.icol(1), marker='o', linestyle='', ms=12, label=name)
ax.legend()
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
#clf = svm.LinearSVC()
clf = svm.SVC()
#clf = RandomForestClassifier(n_estimators=50)

clf.fit(feature_vector_list, solutions.values)
predicted = clf.predict(feature_vector_list)

error = sum(predicted != solutions) / float(solutions.size)
print("Error: " + str(int(10000*error) / 100.0) + "%")
```

```python
other_solutions = pd.read_csv("./input_data/training_solutions_rev1.csv", index_col="GalaxyID")
other_solutions = other_solutions.ix[:,"Class1.1"]

other_solutions = other_solutions.drop(solutions.index)
index = random.sample(other_solutions.index, len(solutions.index))
other_solutions = other_solutions[index]

other_solutions_temp = other_solutions.copy()
other_solutions[other_solutions_temp > 0.5] = 1
other_solutions[other_solutions_temp < 0.5] = 0
```

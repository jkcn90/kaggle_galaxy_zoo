## Part 0: Explore Data

### Area Ratios and Hue Lightness Saturation

```python
disp = cv.Display(displaytype='notebook')
```

```python
img = cv.Image("./input_data/images_training_rev1/100023.jpg")
#img = img.scale(0.3)
```

```python
blobs = img.findBlobs()
blobs.draw()
img.save(disp)
```

```python
blobs = img.findBlobs()
blobs = blobs.filter(blobs.area() == max(blobs.area()))
cropped_image = blobs.crop()[0]
cropped_image = cropped_image.toHLS()
cropped_image.save(disp)
```

#### Sample Features (Physical)

```python
feature_area = (blobs.area() / cropped_image.area())[0]
(feature_hue, feature_lightness, feature_saturation) = cropped_image.meanColor()
feature_array = [feature_area, feature_hue, feature_lightness, feature_saturation]
print(feature_array)
```

#### Sample Features (Raw)

```python
cropped_image = cropped_image.toRGB()
cropped_image = cropped_image.resize(5, 5)
raw_array = cropped_image.getNumpyCv2()
plt.imshow(raw_array)
```

```python
plt.imshow(raw_array[:,:,0])
```

```python
feature_area = raw_array[:,:,0].flatten()
print(feature_area)
```

# Part 1: Import and Preprocess Data

```python
from sklearn import preprocessing
data = GalaxyData()
[training_features, training_solutions, validation_features, validation_solutions] = data.split_training_and_validation_data()
training_solutions.columns = ['Class']
features_and_solutions = training_features.join(training_solutions)
```

```python
groups = features_and_solutions.groupby('Class')

(fig, ax) = plt.subplots()
ax.margins(0.05)
for name, group in groups:
    ax.plot(group.lightness, group.saturation, marker='o', linestyle='', ms=12, label=name)
ax.legend()
plt.xlabel('hue')
plt.ylabel('saturation')
```

```python
import exploration
import feature_extraction
from galaxy_plotting import plot_projection

from sklearn import (manifold, datasets, decomposition, ensemble, lda, random_projection)

%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
solutions = exploration.get_reduced_solutions(upper_threshold=1, lower_threshold=1)
(X, y) = exploration.extract_features(feature_extraction.raw, solutions.index)
```

```python
rp = random_projection.SparseRandomProjection(n_components=2)
X_random = rp.fit_transform(X)
plot_projection(X_random, y)
```

```python
X_pca = decomposition.PCA(n_components=2, whiten=True).fit_transform(X)
plot_projection(X_pca, y)
```

```python
X_iso = manifold.Isomap(n_neighbors=5, n_components=6).fit_transform(X)
plot_projection(X_iso, y)
```

```python
clf = manifold.LocallyLinearEmbedding(n_neighbors=5, n_components=6, eigen_solver='dense', )
X_lle = clf.fit_transform(X)
plot_projection(X_lle, y)
```

```python
clf = manifold.MDS(n_components=2, n_init=1, max_iter=100)
X_mds = clf.fit_transform(X)
plot_projection(X_mds, y)
```

```python
hasher = ensemble.RandomTreesEmbedding(n_estimators=200, random_state=0,
                                       max_depth=5)
X_transformed = hasher.fit_transform(X)
pca = decomposition.TruncatedSVD(n_components=2)
X_reduced = pca.fit_transform(X_transformed)
plot_projection(X_reduced, y)
```

```python
embedder = manifold.SpectralEmbedding(n_components=2, random_state=0,
                                      eigen_solver="arpack")
X_se = embedder.fit_transform(X)
plot_projection(X_se, y)
```
